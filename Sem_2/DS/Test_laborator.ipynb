```python
import numpy as np
import matplotlib.pyplot as plt

# Differential equation
def f(x, y):
    return x * np.exp(x) - y

# Exact solution for comparison
def exact_solution(x):
    return x * np.exp(x)

# Euler's method implementation
def euler_method(f, x0, y0, x_end, h):
    x_values = np.arange(x0, x_end + h, h)
    y_values = [y0]
    y = y0
    for x in x_values[:-1]:
        y += h * f(x, y)
        y_values.append(y)
    return x_values, y_values

# Initial conditions
x0 = 1
y0 = np.exp(1)
x_end = 2
h = 0.1

# Get numerical solution using Euler's method
x_values, y_values = euler_method(f, x0, y0, x_end, h)

# Get exact solution values
exact_values = [exact_solution(x) for x in x_values]

# Calculate the errors
errors = [exact - approx for exact, approx in zip(exact_values, y_values)]

# approximation and exact solution
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, 'b-', label='Euler Approximation')
plt.plot(x_values, exact_values, 'r--', label='Exact Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method vs Exact Solution")
plt.legend()
plt.grid(True)
plt.show()

# errors 
plt.figure(figsize=(10, 6))
plt.plot(x_values, errors, 'm-', label='Errors', color='pink')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Error in Euler Approximation')
plt.legend()
plt.grid(True)
plt.show()
```
