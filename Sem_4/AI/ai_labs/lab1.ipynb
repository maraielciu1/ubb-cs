```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

# Create a 1x2 grid of plots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y, label="sin(x)")
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, label="cos(x)", color='r')
plt.title("Cosine Wave")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.legend()

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 25, 20, 25]  # Percentages
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Custom colors
explode = (0.1, 0, 0, 0)  # Explode the first slice (Category A)

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',  # Show percentages with 1 decimal place
    startangle=90,      # Rotate the start of the pie chart
    shadow=True         # Add a shadow effect
)

plt.title("Pie Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(44)

t = np.arange(0, 30, 0.01)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10 Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, NFFT=256, Fs=1. / 0.01)
axs[1].set_ylabel('Coherence')

plt.show()
```

```python
def spinner(probabilityDistribution):
    return np.random.choice(range(len(probabilityDistribution)), p=probabilityDistribution)
```

```python
probabilityDistribution = np.array([0.1, 0.2, 0.3, 0.4])
spinner(probabilityDistribution)
```

```python
def get_angles(angles):
    return np.array(angles) / sum(angles)
```

```python
angles = np.array([45, 85, 70, 160])
colors = ["blue", "red", "yellow", "green"]
colors[spinner(get_angles(angles))]
```

```python
import numpy as np
R = 0.3
l = 1

def generate_montecarlo_trials(n):
    x = np.random.uniform(-l/2, l/2, n)
    y = np.random.uniform(-l/2, l/2, n)
    distances = np.sqrt(x**2 + y**2)
    is_inside = distances < R
    summ = np.sum(is_inside)
    return summ/n
    
trial_numbers = [10, 100, 200, 3000, 10000, 45000, 6000000]
for n in trial_numbers:
    print("n = %d, estimate = %f" % (n, generate_montecarlo_trials(n)))
```
